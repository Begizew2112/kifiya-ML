```python
df1 =  df[df['GHI']>0]
df1.shape
```

```python
df2 =  df1[df1['DNI']>0]
df2.shape
```

```python
df3 =  df2[df2['DHI']>0]
df3.shape 
```

```python
df3.describe()
```

```python

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df3['ModA'].quantile(0.25)
Q3 = df3['ModA'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df3[(df3['ModA'] < Q1 - 1.5 * IQR) | (df3['ModA'] > Q3 + 1.5 * IQR)]
print(outliers) # the result show empty data frame means that there is no outliers
```

```python
Q1 = df3['ModB'].quantile(0.25)
Q3 = df3['ModB'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df3[(df3['ModB'] < Q1 - 1.5 * IQR) | (df3['ModB'] > Q3 + 1.5 * IQR)]
print(outliers) # the result show empty data frame means that there is no outliers
```

```python
Q1 = df3['WS'].quantile(0.25)
Q3 = df3['WS'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df3[(df3['WS'] < Q1 - 1.5 * IQR) | (df3['WS'] > Q3 + 1.5 * IQR)]
print(outliers) # the result show there  is an outliers so we drop it
```

```python

# Determine the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Drop outliers
df3_cleaned = df3[(df['WS'] >= lower_bound) & (df3['WS'] <= upper_bound)]

# Verify the result
print(df3_cleaned)
```

```python
Q11 = df3_cleaned['WSgust'].quantile(0.25)
Q33 = df3_cleaned['WSgust'].quantile(0.75)
IQR = Q33 - Q11

# Identify outliers
outliers = df3_cleaned[(df3_cleaned['WSgust'] < Q11 - 1.5 * IQR) | (df3_cleaned['WSgust'] > Q33 + 1.5 * IQR)]
print(outliers) #the result shows there is still anoutlier in this columns , we drop it 
```

```python
# Determine the bounds for outliers
l_bound = Q11 - 1.5 * IQR
up_bound = Q33 + 1.5 * IQR

# Drop outliers
df3_cleaned1 = df3_cleaned[(df['WSgust'] >= l_bound) & (df3_cleaned['WSgust'] <= up_bound)]

# Verify the result
print(df3_cleaned1)
```
